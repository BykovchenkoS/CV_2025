```python
import cv2
import matplotlib.pyplot as plt
from gauss_filter import GaussianFilter
```

```python
gauss_filter = GaussianFilter()

kernel_sizes = [9,15,21]
sigma = 20
```

```python
results = gauss_filter.compare_performance('input.jpg', kernel_sizes, sigma)
```

```python
original_image = cv2.imread('input.jpg')
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
gauss_filter.visualize_results(original_image, results)
```

```python
sizes = list(results.keys())
native_times = [results[size]['native_time'] for size in sizes]
opencv_times = [results[size]['opencv_time'] for size in sizes]

plt.figure(figsize=(10, 6))
plt.plot(sizes, native_times, 'o-', label='Native Implementation', linewidth=2)
plt.plot(sizes, opencv_times, 's-', label='OpenCV Implementation', linewidth=2)
plt.xlabel('Kernel Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Performance Comparison: Gaussian Filter')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('performance_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
