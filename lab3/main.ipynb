```python
from image_classifier import ImageClassifier, DatasetManager
```

```python
dm = DatasetManager("dataset", image_resize=(64, 64))
train_loader, test_loader, val_loader = dm.create_dataloaders(batch_size=32)
```

```python
classifier = ImageClassifier(classes=dm.classes)
classifier.set_dataloaders(train_loader, test_loader, val_loader)
```

```python
classifier.fit(epochs=10)
```

```python
classifier.evaluate_metrics()
```

```python
classifier.predict_and_visualize(n=6)
```

```python
classifier.save("model.pt")
```
